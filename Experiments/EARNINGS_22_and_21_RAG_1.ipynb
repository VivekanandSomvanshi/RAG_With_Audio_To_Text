```python
import openai
def transcribe_audio(audio_file):
    response = openai.Audio.transcribe(
        file=audio_file,
        model="whisper-1"  # Use the appropriate model
    )
    return response['text']

transcripts = []
for i in range(len(ds['test'])):
    audio_data = ds['test'][i]['audio']['array']
    transcripts.append(transcribe_audio(audio_data))
```

```python
def summarize_text(text):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=f"Summarize the following text:\n{text}",
        max_tokens=150  # Adjust according to your needs
    )
    return response['choices'][0]['text'].strip()

summaries = [summarize_text(transcription) for transcription in transcripts]
```

#### GEMINI_1.5_FLASH

```python
import torch
from transformers import pipeline
import pandas as pd
from datasets import load_dataset

# Load the dataset
ds = load_dataset("distil-whisper/earnings21")

# Initialize the Gemini 1.5 Flash model for transcription and summarization
model_id = "google/gemini-1.5-flash"
pipe = pipeline("text-generation", model=model_id)

def process_audio(example):
    # Convert audio to text using Gemini 1.5 Flash
    audio_input = example["audio"]
    transcription = pipe(audio_input, task="text-generation")["generated_text"]

    # Summarize the transcription
    summary = pipe(transcription, task="text-generation", max_length=100)["generated_text"]

    return {"file_id": example["file_id"], "summary": summary}

# Apply the processing function to the dataset
processed_data = ds["test"].map(process_audio, batched=True)

# Convert to a pandas DataFrame
df = pd.DataFrame(processed_data)

print(df)
```
